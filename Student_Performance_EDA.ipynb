```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data=pd.read_csv("student-mat.csv",sep=';') 
print(data.head())
```

```python
print(data.isnull().sum())
data.dropna()
print(data)
```

# Data Vizualization

```python
plt.figure(figsize=(10,5)) 
sns.histplot(data['G3'],kde=True,bins=15,color="purple",edgecolor="Black")
plt.xlabel("Final Grade",fontsize=14)
plt.ylabel("Frequency",fontsize=14)
plt.title("Distribution of final Grades",fontsize=20)
plt.show()
```

# correlaton heatmap

```python
plt.figure(figsize=(10,5)) 
numeric_data=data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap of feature")
plt.show()
```

```python
plt.figure(figsize=(10,5)) 
sns.boxplot(x=data['studytime'],y=data['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Effect of study time on final grade")
plt.show()
```
